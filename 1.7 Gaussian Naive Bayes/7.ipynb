```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv('./ionosphere.csv')
df.head()
```

```python
df['class'].unique()
```

```python
df['class'].replace({'g': '0', 'b': '1'}, inplace = True, regex = True)
df.dropna(subset = ["class"], inplace=True)
df.head()
```

```python
x = df.values[:, 1:35]
y = df.values[:, 35]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 100)
```

```python
clf = GaussianNB()
```

```python
clf.fit(X_train, y_train)
yprediction = clf.predict(X_test)
```

```python
print('Resultado previsto:\n', yprediction, '\n')
print('Resultado atual:\n', y_test, '\n')
print('PrecisÃ£o do modelo de Ã¡rvore de decisÃ£o ID3 para estes dados: ', accuracy_score(y_test, yprediction)*100)
```
